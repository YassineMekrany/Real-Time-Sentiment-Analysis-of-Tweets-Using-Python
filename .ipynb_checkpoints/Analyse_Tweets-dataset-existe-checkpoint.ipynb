```python
#entrer un dataset dans MongoDB
import pandas as pd
from pymongo import MongoClient

# Charger le dataset à partir du fichier CSV
file_path = 'twitter_training.csv'
data = pd.read_csv(file_path)

# Connexion à MongoDB (assurez-vous que votre serveur MongoDB est en cours d'exécution)
client = MongoClient('mongodb://localhost:27017/')  # Modifier l'URL de connexion si nécessaire

# Sélectionner ou créer une base de données dans MongoDB
db = client['Tweets']

# Sélectionner une collection dans la base de données MongoDB
collection = db['Tweets_collection']   

# Convertir le DataFrame en une liste de dictionnaires (une ligne par document)
data_dict = data.to_dict(orient='records')

# Insérer les documents dans la collection MongoDB
result = collection.insert_many(data_dict)

# Afficher le nombre de documents insérés
print(f"{len(result.inserted_ids)} documents insérés dans la collection '{collection.name}'")
```

```python
#extraire dataset de MongoDB
import pandas as pd
from pymongo import MongoClient

# Connexion à MongoDB (assurez-vous que votre serveur MongoDB est en cours d'exécution)
client = MongoClient('mongodb://localhost:27017/')  

# Sélectionner la base de données dans MongoDB
db = client['Tweets']  # Remplacez 'my_database' par le nom de votre base de données

# Sélectionner la collection contenant les données
collection = db['Tweets_collection']  # Remplacez 'my_collection' par le nom de votre collection

# Récupérer les documents de la collection MongoDB sous forme de liste de dictionnaires
data_list = list(collection.find())

# Créer un DataFrame pandas à partir des données récupérées
data = pd.DataFrame(data_list)

# Afficher les premières lignes du DataFrame pour vérification
data.head()
```

```python
#Affichage simple

from textblob import TextBlob
import pandas as pd

# Supposons que vos textes sont dans une colonne nommée 'texte'
textes = data['im getting on borderlands and i will murder you all ,']

# Analyse des sentiments pour chaque texte
sentiments = []
for texte in textes:
    blob = TextBlob(str(texte))  # Convertir en chaîne au cas où
    sentiment = blob.sentiment.polarity  # Récupérer la polarité du sentiment
    # Déterminer le sentiment en fonction de la polarité
    if sentiment > 0:
        sentiments.append('Positif')
    elif sentiment == 0:
        sentiments.append('Neutre')
    else:
        sentiments.append('Négatif')

# Ajouter les résultats d'analyse à votre dataset
data['sentiment'] = sentiments

# Afficher les premières lignes du dataset avec les résultats d'analyse
data.head()
```

```python
#Affichage en utilisant le diagramme en secteurs

import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# Supposons que vos textes sont dans une colonne nommée 'texte'
textes = data['im getting on borderlands and i will murder you all ,']

# Analyse des sentiments pour chaque texte
sentiments = []
for texte in textes:
    blob = TextBlob(str(texte))  # Convertir en chaîne au cas où
    sentiment = blob.sentiment.polarity  # Récupérer la polarité du sentiment
    # Déterminer le sentiment en fonction de la polarité
    if sentiment > 0:
        sentiments.append('Positif')
    elif sentiment == 0:
        sentiments.append('Neutre')
    else:
        sentiments.append('Négatif')

# Ajouter les résultats d'analyse à votre dataset
data['sentiment'] = sentiments

# Compter le nombre de chaque type de sentiment
sentiment_counts = data['sentiment'].value_counts()

# Créer un diagramme en secteurs (pie chart)
plt.figure(figsize=(8, 6))
plt.title('Répartition des sentiments')
colors = ['green', 'orange', 'red']
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)
plt.axis('equal')  # Assure que le diagramme en secteurs est un cercle
plt.show()
```

```python
#Affichage en utilisant le diagramme à barres

import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt
import seaborn as sns

# Supposons que vos textes sont dans une colonne nommée 'texte'
textes = data['im getting on borderlands and i will murder you all ,']

# Analyse des sentiments pour chaque texte
sentiments = []
for texte in textes:
    blob = TextBlob(str(texte))  # Convertir en chaîne au cas où
    sentiment = blob.sentiment.polarity  # Récupérer la polarité du sentiment
    # Déterminer le sentiment en fonction de la polarité
    if sentiment > 0:
        sentiments.append('Positif')
    elif sentiment == 0:
        sentiments.append('Neutre')
    else:
        sentiments.append('Négatif')

# Ajouter les résultats d'analyse à votre dataset
data['sentiment'] = sentiments

# Compter le nombre de chaque type de sentiment
sentiment_counts = data['sentiment'].value_counts()

# Créer un diagramme à barres (bar chart)
plt.figure(figsize=(8, 6))
sns.countplot(x='sentiment', data=data, palette='pastel')
plt.title('Répartition des sentiments')
plt.xlabel('Sentiment')
plt.ylabel('Nombre de tweets')
plt.show()
```
