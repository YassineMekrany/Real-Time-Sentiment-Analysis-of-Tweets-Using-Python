```python
import pandas as pd
from pymongo import MongoClient

# Connexion à la base de données MongoDB
client = MongoClient('localhost', 27017)  # Assurez-vous de spécifier les détails de votre serveur MongoDB si nécessaire
db = client['DB']  # Utiliser le même cas que dans MongoDB
collection = db['tweets']  # Remplacez 'tweets' par le nom de votre collection

# Chemin vers votre fichier CSV
csv_file = r"C:\Users\hp\Desktop\Master S2\Big_data\projet\Covid-19.csv"

# Lecture du fichier CSV avec pandas en spécifiant l'encodage
data = pd.read_csv(csv_file, encoding='latin1')  # Essayez 'latin1', 'ISO-8859-1', 'Windows-1252', ou un autre encodage si nécessaire

# Convertir les données en dictionnaires pour insertion dans MongoDB
data_json = data.to_dict(orient='records')

# Insérer les données dans MongoDB
collection.insert_many(data_json)

# Confirmer l'importation
print("Les données ont été importées avec succès dans MongoDB.")
```

```python
import stanfordnlp
from pymongo import MongoClient
from langdetect import detect

# Connexion à la base de données MongoDB
client = MongoClient('localhost', 27017)  # Assurez-vous de spécifier les détails de votre serveur MongoDB si nécessaire
db = client['DB']  # Utiliser le même cas que dans MongoDB
collection = db['tweets']  # Remplacez 'tweets' par le nom de votre collection

# Télécharger les modèles pour les langues dont vous avez besoin
stanfordnlp.download('fr')  # Modèle français
stanfordnlp.download('en')  # Modèle anglais
# Ajoutez d'autres langues au besoin

# Charger le modèle par défaut pour la détection de langue
nlp_langdetect = stanfordnlp.Pipeline(processors='tokenize')

# Récupérer les documents depuis MongoDB
documents = collection.find()

# Analyser chaque document
for document in documents:
    # Supposons que votre texte est stocké dans un champ appelé 'texte'
    text = document['texte']
    
    # Détecter la langue du texte
    doc_langdetect = nlp_langdetect(text)
    detected_language = detect(doc_langdetect.sentences[0].text)

    # Charger le modèle approprié en fonction de la langue détectée
    if detected_language == 'fr':
        nlp = stanfordnlp.Pipeline(lang='fr')
    elif detected_language == 'en':
        nlp = stanfordnlp.Pipeline(lang='en')
    # Ajoutez d'autres langues au besoin

    # Analyse grammaticale
    doc = nlp(text)

    # Afficher les informations sur les tokens (mots)
    for i, sentence in enumerate(doc.sentences):
        print(f"Langue détectée: {detected_language}")
        print(f"Phrase {i+1}:")
        for token in sentence.tokens:
            print(f"Mot: {token.text}\tLemma: {token.lemma}\tPoS: {token.pos}\tFeats: {token.feats}")
```
